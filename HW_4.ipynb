```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
plt.rcParams["figure.figsize"] = (20, 15)
```

```python
t = np.linspace(0,np.pi*100,int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
```

### Problem 1

#### 1a) Make a scatter plot (not a line plot) of the first 1000 data points.

```python
plt.plot(t[0:1000], bkgd[0:1000], 'o')
plt.xlabel('t',fontsize=15)
plt.ylabel('Background',fontsize=15)
plt.title('First 1000 Data Points',fontsize=18)
plt.show()
```

#### 1b) Generalize your code so you can make a plot of any X contiguous points (e.g. 2000 points starting at point 500), and produce an example plot of a set of data somethere in the middle of your array.

```python
def plot_section(start, length, x ,y):
    plt.plot(x[start:start+length],y[start:start+length],'o' ,ms=4)
    plt.show
    

plot_section(500,2000,t,bkgd)
plt.xlabel('t',fontsize=18)
plt.ylabel('Background',fontsize=18)
plt.title('2000 Points Starting at 500',fontsize=18)
plt.show()
```

#### 1c) Sometimes you want to sample the data, such as plotting every 100th point. Make a plot of the full data range, but only every 100th point.

```python
plt.plot(t,bkgd,'o',markevery=100,ms=5)
plt.title('Every 100th Point',fontsize=18)
plt.xlabel('t',fontsize=15)
plt.ylabel('Background',fontsize=15)
plt.show()
```

### Problem 2

#### 2a) Make a 2d histogram plot. Read about options, and play with binsize so it is useful (too large a bin and everything washes out, too small and the sample noise dominates). Make sure you use a colorbar.

```python
fig, ax = plt.subplots(1, 1)
plt.rcParams["figure.figsize"] = (15,10)
plt.xlabel('t',fontsize=15)
plt.ylabel('Background',fontsize=15)
h = ax.hist2d(t,bkgd,bins=100, density=False)
plt.xlabel('t',fontsize=15)
plt.ylabel('Background',fontsize=15)
plt.title('2D Histogram of the Background',fontsize=18)
plt.colorbar(h[3], ax=ax)
plt.show()
```

#### 2b) Clearly explain what is being plotted in your plot (e.g. what does each color mean?)

The colors correspond to how frequent a certain value shows up, so if there are multiple occurrences of a value the bar on the right will indicate how often a value shows by a color.

### Problem 3

#### Make an new time variable that folds at a period of  2  (the modulo operator is your friend, variably expressed as mod(), %, or rem()). Then make a scatter plot of all your data, but now folded. (Reminder, for more data you'd never actually want to plot all the points.)

```python
folded_t = t%(2*np.pi)

plt.plot(folded_t,bkgd,'o',ms=3,markevery=1)
plt.title('Scatter plot of the folded backgound')
plt.xlabel('Folded t',fontsize=15)
plt.ylabel('Background',fontsize=15)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
plt.rcParams["figure.figsize"] = (20,10)
h = ax.hist2d(folded_t,bkgd,bins=100, density=True);
plt.title('2D histagrm of the folded backgound')
plt.xlabel('Folded t',fontsize=18)
plt.ylabel('Background',fontsize=18)
plt.colorbar(h[3], ax=ax)
plt.show()
```
